```python
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
url = 'https://www.1mg.com/categories/exclusive/immunity-boosters/vitamin-c-734'
```

```python
html = requests.get(url = url)
```

```python
html.status_code
```

```python
bsobj = soup(html.content, 'lxml')
bsobj
```

```python
product_name = []

for name in bsobj.find_all('div', {'class':'style__pro-title___2QwJy'}):
    product_name.append(name.text.strip())
```

```python
pack_size = []

for size in bsobj.find_all('div', {'class' : 'style__pack-size___2JQG7'}):
    pack_size.append(size.text.strip())  
```

```python
mrp = []

for price in bsobj.find_all('div', {'class' : 'style__price-tag___cOxYc'}):
    mrp.append(price.text.replace('₹','').replace('MRP','').strip())
```

```python
df = pd.DataFrame({'Product Name':product_name,'Pack Size':pack_size,'Price':mrp}) 
```

```python
df
```

```python
df.to_csv('products.csv', index=False, encoding='utf-8')
```
